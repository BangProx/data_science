```python
# 감마 분포에서 k는 모양, Theta는 스케일이다.
# random.gamma(모양, 스케일 = 1.0,크기 = 없음)
import numpy as np
import matplotlib.pyplot as plt
```

```python
g_list = np.random.gamma(1,100,3000)
plt.hist(g_list,bins = 200,label = 'gamma distribution')
plt.legend()
plt.show
print('mean of random number : {}'.format(g_list.mean()))
```

```python
samples = np.random.choice(g_list, 100)
print('mean of samples is {}'.format(samples.mean()))
```

```python
s_mean_prev = []
for i in range(10000):
    np.random.seed(i)
    s_mean_prev.append(np.random.choice(g_list,100).mean())
s_mean = np.array(s_mean_prev)
```

```python
plt.hist(s_mean, bins = 200,label = 's_mean')
plt.legend()
plt.show
```

```python
#standard deviation은 표준 편차
smm = s_mean.mean()
sms = np.std(s_mean)
print('Mean of s_mean is {}'.format(smm))
print('Standard Deviation of s_mean is {}'.format(sms))
```

```python
#normal distribution은 정규 분포
s_mean_norm = np.random.normal(smm,sms,3000)
plt.hist(s_mean_norm, bins = 200, label = 'Normal Distribution of s_mean')
plt.legend()
plt.show()
```

```python
#s_mean의 그래프와 s_mean의 정규 분포의 그래프는 서로 다른 모양이다.
#중심 극한 정리는 동일한 확률 분포를 가진 독립 확률 변수 n개의 분포는 n이 적당히 커진다면
#정규분포에 가까워진다는 이론이다
#이러한 중심 극한 정리에 따르면 두 그래프는 유사할수는 있어도 동일하지는 않다.
```
