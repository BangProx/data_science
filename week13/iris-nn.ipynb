```python
#1.패키지 설정
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#2.데이터 준비
Dataset = pd.read_csv("/kaggle/input/iris/Iris.csv")
Dataset = Dataset.replace(to_replace ="Iris-setosa",          value ="0") 
Dataset = Dataset.replace(to_replace ="Iris-versicolor",      value ="1") 
Dataset = Dataset.replace(to_replace ="Iris-virginica",       value ="2") 
X = np.array(Dataset[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']])                   # Input
Y = np.array(Dataset["Species"])                                                                            # Targets
```

```python
#3.데이터 탐색
#feature별 박스플록 
plt.boxplot([X[:,0],X[:,1],X[:,2],X[:,3]],
            labels = Dataset.columns[1:5])
plt.title('IRIS')
plt.xlabel('Features')
plt.ylabel('Length or Width(cm)')
plt.show()
```

```python
#변수간 산점도
sns.pairplot(Dataset,hue="Species")
```

```python
#4. 학습용 데이터 분리
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state = 100)
```

```python
#5. 데이터 전처리(정규화)
scaleX=MinMaxScaler() # 생성
scaleX.fit(X_train)   # 맞춤
X_train_std = scaleX.transform(X_train) # 변환
X_test_std = scaleX.transform(X_test) # 변환
```

```python
#6. 모델링 (학습)
model = Perceptron(verbose=1) #생성
model.fit(X_train_std,Y_train)
```

```python
#weight와 bias값 확인
print(model.coef_)
print(model.intercept_)
```

```python
#7. 예측 및 검증
y_pred = model.predict(X_test_std)
print('예측값: ', y_pred)
print('실제값: ', Y_test)
print('정확도: %.2f'%accuracy_score(Y_test,y_pred) )
```

```python
import keras
from keras import Sequential
from keras.layers import Dense

Y_train_vec = keras.utils.np_utils.to_categorical(Y_train, num_classes=3)
Y_test_vec = keras.utils.np_utils.to_categorical(Y_test, num_classes=3)

# Create a new Sequential object
model = Sequential()

# Create the input layer, 50 nodes
model.add(Dense(10, input_shape=(4,)))

# Create the hidden layer
model.add(Dense(10, activation="relu"))
          
# Create an output layer, 3 nodes
model.add(Dense(3, activation="softmax"))

# Compile model
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['acc'])

# Train the NN model
model.fit(x=X_train_std, y=Y_train_vec, epochs=100, batch_size=2, validation_split=0.10)
```

```python
y_pred = model.predict(X_test_std)
y_pred_ind = np.zeros(len(y_pred))
for i in range(0,len(y_pred)):
    print('실제값: ', Y_test[i], '  / 예측값: ',np.argmax(y_pred[i]) , ' :  ',y_pred[i])
    y_pred_ind[i] = np.argmax(y_pred[i])
print('정확도: %.2f'%accuracy_score(Y_test,y_pred_ind.astype('int').astype('str') ))
```
