```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.plot([1,2,3,4])
plt.ylabel('ylabel')
plt.xlabel('xlabel')
```

```python
x = np.arange(10)
plt.plot(x**2)
```

```python
plt.plot(x**2)
plt.axis([0,100,0,100])
```

```python
x = np.arange(-20,20)
y1 = 2 * x
y2 = (1/3) * x**2 + 5
y3 = -x**2 - 5
plt.plot(x,y1,'g--',x,y2,'r^-',x,y3,'b*:')
plt.axis([-30,30,-30,30])
```

```python
x = np.linspace(0,np.pi * 2, 100)
y1= np.sin(x)
y2= np.cos(x)
plt.plot(x,y1,'r-',label = 'sin curve')
plt.plot(x,y2,'b:',label = 'cos curve')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.legend()
```

```python
fig, ax = plt.subplots(2,2)
X = np.random.randn(100)
Y = np.random.randn(100)
ax[0,0].scatter(X,Y)
X = np.arange(10)
Y = np.random.uniform(1,10,10)
ax[0,1].bar(X,Y)
X = np.linspace(0,10,100)
Y = np.cos(X)
ax[1,0].plot(X,Y)
Z = np.random.uniform(0,1,(5,5))
ax[1,1].imshow(Z)
```

```python
fig, ax = plt.subplots(2,3)
grid = plt.GridSpec(2,3,wspace = 0.4,hspace = 0.3)
X = np.random.randn(100)
Y = np.random.randn(100)
plt.subplot(grid[0,0]).scatter(X,Y)
X = np.arange(10)
Y = np.random.uniform(1,10,10)
plt.subplot(grid[0,1:]).bar(X,Y)
X = np.linspace(0,10,100)
Y = np.cos(X)
plt.subplot(grid[1,:2]).plot(X,Y)
Z = np.random.uniform(0,1,(5,5))
plt.subplot(grid[1,2]).imshow(Z)
```

```python
N = 30
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2

plt.scatter(x,y,s = area, c = colors, alpha = 0.5)
```

```python
x = np.arange(3)
years = ['2010','2011','2012']
domestic = [6801,7695,8010]
foreign = [777,1046,1681]
plt.bar(x,domestic)
plt.bar(x,foreign)
plt.xticks(x,years)
```

```python
data = [5,4,6,11]
clist = ['cyan','gray','orange','red']
plt.pie(data,autopct = '%.2f',colors = clist)
```
